**Importing Python Libraries**

```python
import pandas as pd
import numpy as np 
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline

import warnings
warnings.filterwarnings('ignore')
```

```python
from matplotlib import pyplot as plt
plt.style.use('ggplot')
```

**Loading the dataset**

```python
ds = pd.read_csv("Data_Science_Jobs_in_India.csv")
```

**Checking the dataset from top as well as bottom**

```python
ds.head()
```

```python
ds.tail()
```

```python
ds.head(30)
```

**Conversion of Salary from US dollars to INR**

```python
salary = ['avg_salary','min_salary','max_salary']
for field in salary:
  ds[field] = (ds[field].apply(lambda x: float(x[0:-1]) * 100000) / 78.0).astype(int)
ds
```

**Graphical Representation of Indian Data Science Salaries**

```python
ax = ds.iloc[0:10,:].plot.bar(x='company_name', y=['avg_salary','min_salary','max_salary'],figsize=(16,10))
ax.legend(["Average Salary", "Minimum Salary","Maximum Salary"]);
ax.set_ylabel("Salary in US dollars")
ax.set_title("Indian Data Science Salaries")
plt.show()
```



**Graphical Representation of Indian Data Science Highest Average Salaries**



```python
ax = ds.sort_values(by='avg_salary').tail(10).plot.bar(x='company_name', y=['avg_salary','min_salary','max_salary'],figsize=(16,10))
ax.legend(["Average Salary", "Minimum Salary","Maximum Salary"]);
ax.set_ylabel("Salary in US dollars")
ax.set_title("Indian Data Science Highest Average Salaries")
plt.show()
```

**Graphical Representation of Indian Data Science  Highest Minimum
Salaries**



```python
ax = ds.sort_values(by='min_salary',ascending=False).head(10).plot.bar(x='company_name', y=['avg_salary','min_salary','max_salary'],figsize=(16,10))
ax.legend(["Average Salary", "Minimum Salary","Maximum Salary"]);
ax.set_ylabel("Salary in US dollars")
ax.set_title("Indian Data Science Highest Minimum Salary")
plt.show()
```

**Graphical Representation of Indian Data Science  Highest Maximum
Salaries**


```python
ax = ds.sort_values(by='max_salary',ascending=False).head(10).plot.bar(x='company_name', y=['avg_salary','min_salary','max_salary'],figsize=(16,10))
ax.legend(["Average Salary", "Minimum Salary","Maximum Salary"]);
ax.set_ylabel("Salary in US dollars")
ax.set_title("Indian Data Science Maximum Salary")
plt.show()
```

**List of Data Science Companies with highest salaries**

```python
top_ds = ds.sort_values(by='max_salary',ascending=False).head(20)[['company_name','max_salary']].reset_index()
top_ds = ds[['company_name','max_salary']]
top_ds.index += 1
top_ds.columns = ["Company", "Maximum Salary"]
top_ds
```
